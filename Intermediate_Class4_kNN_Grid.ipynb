```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing


df = pd.read_csv("https://an-utd-python.s3.us-west-1.amazonaws.com/congress_votes.csv")

# See data details at https://archive.ics.uci.edu/ml/datasets/congressional+voting+records
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split


# Create arrays for the features and the response variable
y = df['party']
X = df.drop('party', axis=1) # features
le = preprocessing.LabelEncoder()
X = X.apply(le.fit_transform) # convert every column to number. we don't keep track of translations or encoders/decoders. if you expecting more data, you should store the transfornmers

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X
```

```python
y         # output variable -> if only one Y, you can leave as categorical            
```

```python
# Create a k-NN classifier with 6 neighbors
# change n_neighbors
knn = KNeighborsClassifier(n_neighbors=5, weights="distance")
```

```python
knn.fit(X_train, y_train) 
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

## Work on the cars mpg dataset

```python
cars = pd.read_csv("https://an-vistra.s3.us-west-1.amazonaws.com/data/auto-mpg.csv")
```

```python
cars.info()
```

```python
X = cars[['cyl', 'displ', 'hp', 'weight', 'accel', 'origin', 'color']]
y = cars['mpg']
```

```python
le = preprocessing.LabelEncoder()
X = X.apply(le.fit_transform) # convert every column to number. we don't keep track of translations or encoders/decoders. if you expecting more data, you should store the transfornmers

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors = 13, weights = "distance")
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import explained_variance_score, r2_score, mean_absolute_error, mean_squared_error
y_pred = knn.predict(X_test)
print(explained_variance_score(y_test, y_pred))
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

## Pipelines

Pipelines create a series of steps that can be run one after the other, without having to wait for individual steps to terminate.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipe = Pipeline([('scaler', StandardScaler()) ,('knn', KNeighborsRegressor(n_neighbors=6))])
# The pipeline can be used as any other estimator
# and avoids leaking the test set into the train set

pipe.fit(X_train, y_train)

pipe.score(X_test, y_test)

```

# GridSearch with Cross-Validation

```python
# i am running on the voting dataset
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

params = {'n_neighbors': [1, 3, 5, 7, 9, 11], 'weights': ['distance', 'uniform'], 'p': [1, 2, 3]}  # i want to try these values of n_neighrs

grid = GridSearchCV(knn, params, cv=10, scoring='accuracy', return_train_score=False)


grid.fit(X_train, y_train)

best_model = grid.best_estimator_
print(grid.best_params_)

print(best_model.score(X_test, y_test))
# find best model score

# print(grid.cv_results_)
```

Take the Car Dataset and apply a knn model to predict the origin of car

```python
cars
```

```python
X = cars[['mpg', 'cyl', 'displ', 'hp', 'weight', 'accel', 'size']]
y = cars['origin']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

We want to first: standardize data using MinMaxScaler

We want to fit a knn classifier model, and i want to try following parameters:
n_neighbirs = [3, 5, 7, 9] 

and knn__weights': ['uniform', 'distance']}]

Find best model parameters and performance of best model on the test part

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV

knn_pipe = Pipeline([('mms', MinMaxScaler()),
                     ('knn', KNeighborsClassifier())])


params = [{'knn__n_neighbors': [3, 5, 7, 9, 11, 13],
         'knn__weights': ['uniform', 'distance']}]

gs_knn = GridSearchCV(knn_pipe,
                      param_grid=params,
                      scoring='accuracy',
                      cv=5)
gs_knn.fit(X_train, y_train)
print(gs_knn.best_params_)
# find best model score

print(gs_knn.score(X_train, y_train))
```

```python
best_model = gs_knn.best_estimator_
y_pred = best_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_pipe = Pipeline([('mms', MinMaxScaler()),
                     ('dt', DecisionTreeClassifier())])
params = [{'dt__max_depth': [3, 5, 7, 9],
         'dt__min_samples_leaf': [3, 5],
         'dt__min_samples_split': [5, 8, 10, 20],}]
         
gs_dt = GridSearchCV(dt_pipe,
                      param_grid=params,
                      scoring='accuracy',
                      cv=5)
gs_dt.fit(X_train, y_train)
print(gs_dt.best_params_)
# find best model score
print(gs_dt.score(X_train, y_train))
```

```python
best_model = gs_dt.best_estimator_
y_pred = best_model.predict(X_test)

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
